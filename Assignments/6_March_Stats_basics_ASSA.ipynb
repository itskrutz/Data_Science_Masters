Q1. What is Statistics?
Statistics is the branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data. It provides tools and methods for making sense of complex data sets and for making decisions based on data.


Q2. Define the different types of statistics and give an example of when each type might be used.
Statistics can be broadly divided into two types: Descriptive Statistics and Inferential Statistics.

Descriptive Statistics:
Definition: Descriptive statistics involve summarizing and organizing data so that it can be easily understood. This includes calculating measures such as mean, median, mode, range, and standard deviation.
Example: A teacher calculating the average score of students in a class after an exam to summarize the overall performance.

Inferential Statistics:
Definition: Inferential statistics involve making predictions or inferences about a population based on a sample of data taken from that population. This includes hypothesis testing, regression analysis, and confidence intervals.
Example: A pharmaceutical company testing a new drug on a sample of patients to infer the drugâ€™s effectiveness for the 
entire population.

Q5. Explain the concept of levels of measurement and give an example of a variable for each level.
Levels of measurement describe the nature of information within the values assigned to variables and dictate the types of statistical analyses that can be performed. The four levels are:

Nominal:
Definition: Categorizes data without any order.
Example: Types of fruit (apple, banana, cherry).

Ordinal:
Definition: Categorizes data with a meaningful order but without a fixed interval between categories.
Example: Movie ratings (poor, fair, good, excellent).

Interval:
Definition: Data with meaningful intervals between values but no true zero point.
Example: Temperature in Celsius.

Ratio:
Definition: Data with a meaningful zero point and meaningful intervals between values.
Example: Weight (0 kg means no weight).

Understanding the level of measurement is crucial because it determines the appropriate statistical techniques for data analysis. For instance, you cannot calculate the mean of nominal data, and using inappropriate statistical methods can lead to incorrect conclusions.

Example:
If you have ordinal data like customer satisfaction ratings (poor, fair, good, excellent), calculating the average would not make sense, but you can find the median or mode. However, for ratio data like income, calculating the mean, median,

Q8. Which type of plot can be used to display data in terms of range?
A box plot (or box-and-whisker plot) can be used to display data in terms of range. It shows the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, and maximum).

Q10. What are some common measures of central tendency and variability used in statistics? Explain how each measure can be used to describe a dataset.
Measures of Central Tendency:
Mean: The average of the data points.
Use: Provides a central value of the dataset.
Example: Average income of employees in a company.

Median: The middle value of the data points when ordered.
Use: Useful when the dataset contains outliers.
Example: Median home prices in a city.

Mode: The most frequently occurring value in the dataset.
Use: Indicates the most common value.
Example: Most common shoe size sold in a store.

Measures of Variability:
Range: The difference between the highest and lowest values.
Use: Indicates the spread of the dataset.
Example: Range of temperatures recorded in a month.

Variance: The average squared deviation from the mean.
Use: Measures the spread of the data points.
Example: Variance in students' test scores.

Standard Deviation: The square root of the variance.
Use: Indicates how much the data points deviate from the mean.
Example: Standard deviation of weights of individuals in a population.
